```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
mail_df=pd.read_csv('/content/mail_data.csv')
```

```python
mail_df.head()
```

```python
mail_df.isnull().sum()
```

```python
mail_data=mail_df.where((pd.notnull(mail_df)),'')
```

```python
mail_data.head()
```

```python
mail_data['Category'].value_counts()
```

Label Encoding

```python
mail_data.loc[mail_data['Category']=='spam','Category']=0
mail_data.loc[mail_data['Category']=='ham','Category']=1
```

```python
mail_data.head()
```

```python
#seperating the features and targets
X=mail_data.drop(columns='Category',axis=1)
Y=mail_data['Category']
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

Feature Extraction

```python
#converting text into numerical data
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
X_train_features=feature_extraction.fit_transform(X_train['Message'])
X_test_features=feature_extraction.transform(X_test['Message'])

#converting Y_train and Y_test to int
Y_train=Y_train.astype('int')
Y_test=Y_test.astype('int')
```

```python
print(X_train_features)
```

```python
print(X_test_features)
```

Training the ML Model

Logistic Regression

```python
Model=LogisticRegression()
Model.fit(X_train_features,Y_train)
```

```python
Y_train
```

Evaluating our trained mode

```python
#Prediction on our trained data
prediction_on_training_data=Model.predict(X_train_features)
accuracy_on_training_data=accuracy_score(Y_train,prediction_on_training_data)
accuracy_on_training_data
```

```python
prediction_on_testing_data=Model.predict(X_test_features)
accuracy_on_testing_data=accuracy_score(Y_test,prediction_on_testing_data)
accuracy_on_testing_data
```

```python
input_data=['Im going to try for 2 months ha ha only joking']

input_data_features=feature_extraction.transform(input_data)

prediction=Model.predict(input_data_features)

if prediction[0]==0:
  print('Spam')
else:
  print('Ham')

```
